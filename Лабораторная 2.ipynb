```python
def Seidel(A, b , x ,N  ):
    
    xprev = [0.0 for i in range(N)]
    for q in range(10):
        for j in range(N):
            summ = 0.0
            for k in range(N):
                if (k != j):
                    summ = summ + A[j][k] * x[k]
            x[j] = (b[j] - summ) / A[j][j]   
        nev =   [0.0 for i in range(N)]
        modul = 0
        for j in range(N):
            for i  in range(N):
                nev[j] += A[j][i]*x[i]
            nev[j]=nev[j]-b[j]
            modul+=nev[j]*nev[j]
        modul = modul**(0.5)    
        print ("Невязка на итерации номер ", q ,"- это ", modul)
        if ((modul) < 0.00001):
            for i in range(N):
                print(x[i])
            break
    return(0)    
```

# Метод Гаусса с выбором главного элемента

```python
def Gauss_gl_elem(A,b,N):
    column = 0
    answer = [0.0 for i in range(N)]
    while (column < N) :
        curr_row = -1
        for r in range(column, N):
            if ((curr_row == -1) or abs(A[r][column]) > abs(A[curr_row][column])):
                 curr_row = r
        if curr_row != column:
                A[curr_row] , A[column] =  A[column] , A[curr_row]
                b[curr_row] , b[column] =  b[column] , b[curr_row]
        
        A[column] = [i / (A[column][column]) for i in A[column]]
        b[column] /= (A[column][column])
   
        for r in range(column + 1, N):
            A[r] = [(i + j * (-A[r][column])) for i, j in zip(A[r], A[column])]
            b[r] += b[column] * (-A[r][column])         
        column += 1
    for i in range(N-1 , -1, -1):
        t = sum(x * a for x, a in zip(answer[(i + 1):], A[i][(i + 1):]))
        answer[i] = b[i] - t
    for i in range(N):
        print(x[i])
    return(0)
```
