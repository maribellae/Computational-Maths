```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import math

from sympy import diff, symbols, cos, sin, exp , ln, sqrt
x, y = symbols('x y')
i=1
mass = [0]*21

mass[0]=1
while i< len(mass) :
    mass[i] = mass[i-1]/2
    i = i+1
```

```python
#объявляю функции дифференцирования
def df1(x,h):
    Df1 = (f(x+h)-f(x))/(h)
    return (Df1)
def df2(x,h):
    Df2 = (f(x)-f(x-h))/(h)
    return(Df2)
def df3(x,h):
    Df3 = (f(x+h)- f(x-h))/(2*h)
    return(Df3)
def df4(x,h):
    Df4 = (4/3) * ( (f(x+h)-f(x-h))/(2*h)) - (1/3)*( (f(x+(2*h))-f(x-(2*h)))/(4*h))
    return(Df4)
def df5(x,h):
    Df5= (3/2) * ( (f(x+h)-f(x-h))/(2*h)) - (3/5)*( (f(x+(2*h))-f(x-(2*h)))/(4*h))+ (1/10)*( (f(x+(3*h))-f(x-(3*h)))/(6*h))
    return(Df5)
```

```python
def f(x):
    f = sin(x*x)
    return f

def ff(x):
    ff = 2*x*cos(x**2)
    return ff
```

```python
def machine(method): 
    final = [0]* (len(mass))
   
    k = ff(1)
    
    i = 0
    while i< len(mass) :
        final[i]= method(1,mass[i]) - k
        i = i+1     
      
    return(final)

def pl(str):
    figure = plt.figure(figsize=(10,7))
    axes = figure.add_subplot ()

    axes.plot(mass,machine(df1) , 'r', label ="df1")
    axes.plot(mass,machine(df2) , 'g', label ="df2")
    axes.plot(mass,machine(df3), 'y', label ="df3")
    axes.plot(mass, machine(df4), 'pink', label ="df4")
    axes.plot(mass, machine(df5), 'orange', label ="df5")
    plt.legend()
    axes.set_xscale ('symlog')
    axes.set_yscale ('symlog')
    plt.xlabel("log (h) ")
    plt.ylabel("log (  df (h) - df_real  ) ")
    plt.title(str)
    plt.show() 
    return(0)

pl("sin(x*x)")
```
