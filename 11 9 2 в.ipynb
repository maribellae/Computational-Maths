```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(v): 
    f = np.ndarray(4);
    f[0] = -np.cos(v[3])*v[1];
    f[1] = v[0];
    f[2] = -np.sin(v[3])
    f[3] = v[2]
    return f

def MRK(v0,n, distance):
   
   
    h = distance/n
    for i in range(n):
        v = v0 + h*f( v0 + 0.5*h*f(v0))
        v0 = v;
    return v[1], v[3]

def MRK_final(v0,n , distance):

    
    y = []
    h =  distance/n
    for i in range(n):
        v = v0 + h*f( v0 + 0.5*h*f(v0))
        v0 = v;
        y.append(v0[3])
    return v[1], v[3] , y

def Newton(err , distance , n):
    a0= 0
    z=y=100 
    a = 0
    while(( ((y - np.pi)/z )*((y - np.pi)/z ))>err*err):
        
        v = np.array([1,0,a0,0])
        z,y= MRK(v,n, distance)
        a = a0 - (y-np.pi)/z
        
        a0=a
 
    z,y,res_y = MRK_final(v,n, distance)
    return y, res_y, a

def results(err):
    distances = [0.5 , 1,2 ]
    N = [1000,1000, 1000]
    for  i in range(3):
        y,res_y, alpha = Newton(0.0001 , distances[i] , N[i]);
        plt.plot(np.linspace(0,distances[i],len(res_y)), res_y);
        print(" THE ALPHA IS ", alpha ,", THE VALUE IS", y )

  
```

```python
results(0.01)
```
